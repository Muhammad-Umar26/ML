```python
import pandas as pd
import seaborn as sns
```

```python
# B)
df = pd.read_csv('cancer_patients.csv')
df.head()
```

```python
# C)
df['Level'].unique()
```

```python
# D, E)
print(len(df.columns))
display(df.head(50))
display(df.tail(30))
print(df.info())
print(df.shape)
```

```python
# F)
print(df['Level'].value_counts())
sns.barplot(x=['Low', 'Medium', 'High'], y=df['Level'].value_counts())
```

```python
# G)
df1 = df.loc[550:901, 'Gender':'Coughing of Blood']
df1.reset_index(drop=True, inplace=True)

print('--------------------Statisitical Analysis----------------------\n')
for col in df1.columns:
    print(f'{col.upper()}:')
    print(f'Mean: {df1[col].mean()}')
    print(f'Median: {df1[col].median()}')
    print(f'Mode: {df1[col].mode()[0]}')
    print(f'Standard Deviation: {df1[col].std()}')
    print(f'Minimum: {df1[col].min()}')
    print(f'Maximum: {df1[col].max()}\n')
```

```python
# H)
df2 = df.iloc[-20:, :]
df2 = df2.loc[:, 'Shortness of Breath':'Frequent Cold']
df2.reset_index(drop=True, inplace=True)

print('--------------------Statisitical Analysis----------------------\n')
for col in df2.columns:
    print(f'{col.upper()}:')
    print(f'Mean: {df2[col].mean()}')
    print(f'Median: {df2[col].median()}')
    print(f'Mode: {df2[col].mode()[0]}')
    print(f'Standard Deviation: {df2[col].std()}')
    print(f'Minimum: {df2[col].min()}')
    print(f'Maximum: {df2[col].max()}\n')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cancer_patients.csv')
df.head()
```

```python
# B)
plt.figure(figsize=(18, 12))
sns.heatmap(df.drop(columns=['Patient Id', 'Level']).corr(), annot=True)
plt.title('Correlation Graph', fontsize=24, pad=15)
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age", fontsize=16)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Count', fontsize=15)

# Frequent Cold Distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Frequent Cold'], kde=True)
plt.title("Distribution of Frequent Cold", fontsize=16)
plt.xlabel('Frequent Cold', fontsize=15)
plt.ylabel('Count', fontsize=15)

plt.tight_layout(pad=4)
plt.show()
```

```python
# D)
sns.boxplot(x=df['Obesity'])
plt.show()
```
